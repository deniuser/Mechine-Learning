```python
!pip install pandas
!pip install mlxtend

import pandas as pd
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules

from google.colab import files

# Unggah file
uploaded = files.upload()

# Mendapatkan nama file yang diunggah
file_name = next(iter(uploaded))

# Membaca file Excel ke dalam DataFrame
data = pd.read_excel(file_name)

# Menampilkan beberapa baris pertama dari data untuk memastikan data telah dimuat dengan benar
print("Data preview:")
print(data.head())

# Memeriksa apakah data sudah dalam format biner
if not data.applymap(lambda x: x in [0, 1, True, False]).all().all():
    # Mengonversi data ke one-hot encoding
    transactions = pd.get_dummies(data)
else:
    transactions = data

# Menampilkan data yang sudah diproses
print("\nPreprocessed Data (One-Hot Encoded):")
print(transactions.head())

def get_highest_confidence_rule(transactions, min_support_range, min_confidence_range):
    highest_confidence = 0
    best_rule = None

    # Iterasi melalui berbagai nilai min_support
    for min_support in min_support_range:
        # Menjalankan algoritma FP-Growth untuk menemukan frequent itemsets
        frequent_itemsets = fpgrowth(transactions, min_support=min_support, use_colnames=True)

        # Iterasi melalui berbagai nilai min_confidence
        for min_confidence in min_confidence_range:
            # Menghasilkan association rules dari frequent itemsets
            rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)
            if not rules.empty:
                # Mencari rule dengan nilai confidence tertinggi
                max_conf_rule = rules.loc[rules['confidence'].idxmax()]
                if max_conf_rule['confidence'] > highest_confidence:
                    highest_confidence = max_conf_rule['confidence']
                    best_rule = max_conf_rule

    return best_rule

# Mendefinisikan rentang nilai untuk min_support dan min_confidence
min_support_range = [x / 100 for x in range(10, 91)]
min_confidence_range = [x / 100 for x in range(10, 91)]

# Mendapatkan rule dengan nilai confidence tertinggi
best_rule = get_highest_confidence_rule(transactions, min_support_range, min_confidence_range)

print("\nBest Rule with Highest Confidence:")
print(best_rule)

```
